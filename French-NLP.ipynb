```python
import csv
import spacy
from sklearn.metrics import classification_report

# Charger le modèle spaCy pour le traitement du langage naturel en français
nlp = spacy.load("fr_core_news_sm")

# Chemin vers votre fichier CSV contenant les données de test
url = "datasetFR.csv"

# Liste pour stocker les textes et les étiquettes de votre ensemble de test
textes_test = []
etiquettes_test = []

# Charger les données de test à partir du fichier CSV
with open(url, "r", encoding="utf-8") as fichier_csv:
    lecteur_csv = csv.reader(fichier_csv)
    next(lecteur_csv)  # Ignorer l'en-tête
    for ligne in lecteur_csv:
        texte_test = ligne[0]
        etiquette_test = ligne[1]  # Assurez-vous d'adapter cet index en fonction de votre fichier CSV
        textes_test.append(texte_test)
        etiquettes_test.append(etiquette_test)

# Prédictions sur l'ensemble de test
predictions = []
for texte in textes_test:
    doc = nlp(texte)
    # Votre logique pour prédire une étiquette en fonction du texte
    # Ici, nous supposons que votre modèle prédit une étiquette pour chaque texte
    # Vous devrez remplacer cette logique par la vôtre
    prediction = "LABEL_PREDIT"
    predictions.append(prediction)

# Évaluation de la performance
rapport_classification = classification_report(etiquettes_test, predictions, zero_division=1)

# Affichage du rapport de classification
print(rapport_classification)

```

```python
import pandas as pd
import spacy

# Charger le modèle spaCy pour le français
nlp = spacy.load("fr_core_news_sm")

# Charger votre ensemble de données depuis le fichier CSV
url = "datasetFR.csv"
df = pd.read_csv(url)

# Vérifier les noms de colonnes
print(df.columns)
```

```python
def enlever_stopwords(texte):
    doc = nlp(texte)
    tokens_sans_stopwords = [token.text for token in doc if not token.is_stop]
    return " ".join(tokens_sans_stopwords)

# Utiliser le nom correct de la colonne contenant les textes
# (remplacez "nom_de_la_colonne" par le nom réel de la colonne)
df["texte_sans_stopwords"] = df["Description"].apply(enlever_stopwords)

# Afficher les premières lignes du DataFrame avec les stopwords enlevés
print(df.head())
```
